```python
import os
from langchain_openai import ChatOpenAI
from langchain.agents import initialize_agent, Tool, AgentType

from langchain_community.tools import DuckDuckGoSearchRun
from langchain_community.tools import DuckDuckGoSearchResults
from langchain_community.utilities import DuckDuckGoSearchAPIWrapper
from langchain_community.utilities import SerpAPIWrapper
from langchain_community.utilities import GoogleSerperAPIWrapper
```

```python
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())
# openai_api_key = os.environ["OPENAI_API_KEY"]

serpapi_api_key = os.environ["SERPAPI_API_KEY"]
serper_api_key = os.environ["SERPER_API_KEY"]
# print(serpapi_api_key)
# print(serper_api_key)

MODEL_GPT="gpt-4o-mini"
```

```python
# Create search tools
def create_search_tools():
    tools = []
    
    # OpenAI Search (via DuckDuckGo)
    # ddg_search = DuckDuckGoSearchRun()
    # tools.append(
    #     Tool(
    #         name="OpenAI (via DuckDuckGo) Search",
    #         func=ddg_search.run,
    #         description="Useful for searching the web using OpenAI Search (via DuckDuckGo)"
    #     )
    # )
    
    # SerpAPI (Google Search API) (if API key is available)
    # try:
    #     serpapi_search = SerpAPIWrapper()
    #     tools.append(
    #         Tool(
    #             name="SerpAPI (Google Search API) Search",
    #             func=serpapi_search.run,
    #             description="Useful for searching the web using SerpAPI (Google Search API)"
    #         )
    #     )
    # except Exception as e:
    #     print(f"SerpAPI not configured: {e}")

    # SerperAPI (Google Search API) (if API key is available)
    try:
        serperapi_search = GoogleSerperAPIWrapper()
        tools.append(
            Tool(
                name="SerperAPI (Google Search API) Search",
                func=serperapi_search.run,
                description="Useful for searching the web using SerperAPI (Google Search API)"
            )
        )
    except Exception as e:
        print(f"SerperAPI not configured: {e}")
    
    return tools
```

```python
# Create the agent
def create_search_agent():
    # Initialize the language model
    llm = ChatOpenAI(temperature=0, model=MODEL_GPT)
    
    # Get search tools
    tools = create_search_tools()
    
    # Initialize the agent
    agent = initialize_agent(
        tools, 
        llm, 
        agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, 
        verbose=True,
        max_iterations=3
    )
    
    return agent
```

```python
# Create the agent
agent = create_search_agent()
```

```python
user_query = "Obama's first name, explain this name shortly, in 50 words, answer in english, spanish and czech language, output as json (language, answer)?"

# Run the agent
response = agent.invoke(user_query)
            
# Print the response
print("\nAgent Response:")
print(response)
```
